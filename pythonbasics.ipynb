```python
#6.Write a function to demonstrate the behavior of mutable and immutable arguments.

def modify_values(immutable_arg, mutable_arg):
    immutable_arg += 10
    print("Inside function (immutable):", immutable_arg)
    mutable_arg.append(10)
    print("Inside function (mutable):", mutable_arg)

immutable_value = 5
mutable_value = [1, 2, 3]

print("Before function call (immutable):", immutable_value)
print("Before function call (mutable):", mutable_value)

modify_values(immutable_value, mutable_value)

print("After function call (immutable):", immutable_value)
print("After function call (mutable):", mutable_value)
```
